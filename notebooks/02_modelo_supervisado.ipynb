```python
from src.data_repository import DataRepository
from src.pipeline import DataPipeline
from src.model_factory import ModelFactory
from src.evaluator import ModelEvaluator

# 1. Cargar y procesar datos usando tu Pipeline y semilla fija (reproducibilidad)
repo = DataRepository(raw_path="data/raw/incidencia_delictiva.csv")
df = repo.load_data()

pipeline = DataPipeline(seed=42)
X_train, X_test, y_train, y_test = pipeline.transform(df)

# 2. Crear y entrenar el modelo mediante tu Factory Method
factory = ModelFactory()
model = factory.create_model(model_type="RandomForest", seed=42)
model.fit(X_train, y_train)

# 3. Evaluar y guardar (Persistencia exigida en la rúbrica)
evaluator = ModelEvaluator()
evaluator.generate_report(model, X_test, y_test)

# Guardar el artefacto final
import joblib
joblib.dump(model, "models/modelo_final.joblib")
```
